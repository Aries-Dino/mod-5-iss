```python
import json
import requests
import time

data = []


while len(data) < 100:
    response = requests.get("http://api.open-notify.org/iss-now.json")
    obj = json.loads(response.content)
    data.append(obj)
    time.sleep(10)
    
data
```

```python
with open('iss.json','w') as f:
    json.dump(data, f, indent=4, sort_keys=True)
```

```python
with open('iss.json','r') as f:
    data_final = json.load(f)
```

```python
x = []

for i in range(len(data_final)):
    x_list = data_final[i]["iss_position"]["latitude"]
    x.append(float(x_list))
```

```python
y = []

for i in range(len(data_final)):
    y_list = data_final[i]["iss_position"]["longitude"]
    y.append(float(y_list))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig, axes = plt.subplots(figsize=(12,12))

plt.scatter(x,y)
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Path of ISS every 10 seconds')
```
